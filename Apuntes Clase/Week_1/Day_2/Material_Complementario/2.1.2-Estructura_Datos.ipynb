# Listas

La listas son variables que almacenan arrays, donde podemos almacenar datos de distintos tipos (strings, integers, floats, listas, etc.). Por lo tanto, una lista es una estructura de datos formada por una secuencia ordenada de objetos.

```python
lst = [1,2,4,6,8]
```

```python
print(lst, type(lst))
```

```python
# si queremos saber cual es el tipo de cada uno de los elementos de nuestra lista
for i in lst:
    print(type(i))
```

```python
for chorizo in lst:
    print(type(chorizo)) # al iterable le podemos poner el nombre que queramosa
```

```python
lst2 = ["hola", 23456, 19.09, 'Adios'] 
```

```python
print(lst2, type(lst2))
```

```python
# si queremos saber la longitud de nuestra lista, o lo que es lo mismo, el nmero de elementos que tiene
print("lst tiene", len(lst), "elementos")
print("lst2 tiene", len(lst2), "elementos")
```

```python
# tambin podemos sacar los valores mximos y mnimos de nuestra lista
#  NOTA: esto solo se podr hacer si todos los elementos de la lista son numricos

print("el mximo es:", max(lst))
print("el mnimo es:", min(lst))
```

```python
lst
```

```python
# tambin podemos chequear si un nmero dado est incluido en nuestra lista:

print(5 in lst)
print(1 in lst)
```

```python
5 not in lst
```

# Diccionarios 

Los diccionarios en Python nos permiten almacenar una serie de mapeos entre dos conjuntos de elementos, llamados keys and values (Claves y Valores).

- Todos los elementos en el diccionario se encuentran encerrados en un par de llaves `{}`.


- Cada elemento en un diccionario contiene una clave (*key*) y un valor (*value*) - es decir un par de clave-valor.



- Cada par de clave-valor es denominado como elemento (item).


- Puede almancenar cualquier tipo de tipo o estructura de datos (*integers*, *floats*, *booleanos*, listas, etc.). 


- Son **mutables** y **sin orden**.


- Las keys son **nicas**. 

-----
**Cmo podemos crear diccionarios?**

- Usando el mtodo `dict()`


- Usando las `{}` y definiendo nosotros mismos las `keys` y los `values`

```python
# usando dict(). Creamos un diccionario vacio

dict1 = dict()
dict1
```

```python
# usando las {} creamos un diccionario vaco
dict2 = {}
dict2
```

```python
# usando las {}
dict2 = {'python': 27, 'plone': 51}
dict2
```

```python
dict3 = {27:'python', 51:'plone'}
dict3
```

```python
# para saber cuantos pares de key:values tenemos usaremos "len" como en las listas
len(dict2)
```

```python
# para conocer las keys que tiene nuestro diccionario
dict2.keys()
```

```python
# para acceder a los values de nuestro diccionario
dict2.values()
```

----
Ya sabemos crear diccionarios, pero... Cmo accedemos a sus elementos si no tienen orden? 

```python
dict2[0]
```

```python
# podemos acceder a los valores de un diccionario a travs de sus keys
dict2["python"]
```

```python
dict2['plone']
```

```python
# para cambiar el value de una key
dict2["python"] = 45
```

## Mtodos de los diccionarios

```python
diccionario = {"Pepe": 45, "Rosa": 30, "Rodrigo": 19, "Irene": 98}
```

`items`: devuelve una lista de pares de diccionarios(key:values), como tuplas

```python
diccionario.items()
```

`get`: busca un elemento a partir de su clave

```python
diccionario.get("Pepe")
```

```python
# si queremos ver los pares de key values usando un for tenemos que usar
## el mtodo .items()
for chorizo, morcilla in diccionario.items():
    print(chorizo, "---->", morcilla)
```

```python
# con este mtodo podemos pasar otro parmetro para que en caso de que no exista la key nos devuelva un valor por defecto
diccionario.get("PEpe", "no existe")
```

`setdefault`: es similar al `get`. En este caso, si la *key* no existe la crea y asigna el value del valor por defecto que indicamos

```python
diccionario.setdefault("PEpe", 'no existe')
```

```python
diccionario
```

```python
type(diccionario['PEpe'])
```
